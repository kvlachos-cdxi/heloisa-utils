```python
# import cartopy.crs as ccrs
```

```python
nc_file = xr.open_dataset('/home/data/mydata/S2B_MSI_2021_12_19_09_29_19_T34TEK_L2R.nc')
nc_file
```

```python
dem = nc_file['pm2p5']
dem
```

```python
aff = dem.rio.transform()
type(aff)
```

```python
aff
```

```python
dem = dem.rio.set_spatial_dims(x_dim='longitude', y_dim='latitude')
dem.rio.crs
```

```python
dem.rio.bounds()
```

```python
dem.rio.nodata
```

```python
dem.rio.resolution()
```

```python
dem.rio.write_crs("EPSG:4326", inplace=True)
```

```python
dem[0].rio.to_raster(r"/home/freekostageo/Desktop/Productivity scripts/dask/data/mydata/medsea_bottomT_raster.tiff")
```
