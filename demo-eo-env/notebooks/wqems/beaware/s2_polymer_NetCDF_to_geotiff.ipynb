```python
nc_file = xr.open_dataset('/home/AOIs/polyphytos/young_research_data/polymer/S2B_MSIL1C_20211219T092309_N0301_R093_T34TEK_20211219T102236/Polymer_19Dec2021.nc')
nc_file
```

```python
nc_file.coords
```

```python
dir(nc_file)
```

```python
band_names = [key for key in nc_file.keys() if ('Rw' in key)]
band_names.extend(['latitude', 'longitude'])
band_names
```

```python
blue = nc_file[band_names]
blue
```

```python
aff = blue.rio.transform()
type(aff)
```

```python
aff
```

```python
blue = blue.rename({'width': 'longitude','height': 'latitude'})
blue
```

```python
help(blue.rio.reproject)
```

```python
blue.rio.set_spatial_dims(x_dim='longitude', y_dim='latitude', inplace=True)
blue.rio.crs
```

```python
blue.rio.bounds()
```

```python
blue.rio.nodata
```

```python
blue.rio.resolution()
```

```python
blue.rio.write_crs('EPSG:4326', inplace=True)
```

```python
# reproject
blue.rio.reproject(dst_crs='EPSG:32634')
```

```python
blue.rio.crs
```

```python
blue.rio.to_raster('/home/AOIs/polyphytos/young_research_data/polymer/S2B_MSIL1C_20211219T092309_N0301_R093_T34TEK_20211219T102236/Polymer_19Dec2021.tif')
```
