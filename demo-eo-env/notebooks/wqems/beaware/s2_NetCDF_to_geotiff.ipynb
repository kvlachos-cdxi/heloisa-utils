```python
nc_file = xr.open_dataset('/mnt/Local_Projects/CERTH/WQeMS/atmospheric_corrections/acolite/S2B_MSI_2021_12_19_09_29_19_T34TEK_L2R.nc')
nc_file
```

```python
crs = nc_file.scene_proj4_string
crs
```

```python
blue = nc_file[['rhos_492', 'rhos_559', 'rhos_665', 'rhos_864']]
blue
```

```python
aff = blue.rio.transform()
type(aff)
```

```python
aff
```

```python
blue = blue.rio.set_spatial_dims(x_dim='x', y_dim='y')
blue.rio.crs
```

```python
blue.rio.bounds()
```

```python
blue.rio.nodata
```

```python
blue.rio.resolution()
```

```python
blue.rio.write_crs(crs, inplace=True)
```

```python
blue.rio.crs
```
